### Figure 5. SparseStack versus SparseIID.  
Relative error  
$$
\frac{\|\mathbf{A} - \hat{\mathbf{A}}\|_{\mathrm{F}}^2}{\|\mathbf{A}\|_{\mathrm{F}}^2}
$$  
for approximations $\hat{\mathbf{A}}$ obtained by the RSVD using a SparseStack, a SparseIID, or a Gaussian test matrix.  

We consider four diagonal input matrices $\mathbf{A} \in \mathbb{R}^{1024 \times 1024}$ from the testbed in Tropp et al. (2017). For each instance, we perform 100 trials. The data markers track the median errors, and shaded regions are bounded by the 10% and 90% quantiles.  

The SparseStack test matrix with row sparsity $\zeta=3$ or $\zeta=4$ consistently matches the accuracy benchmark of a Gaussian test matrix, whereas the SparseIID test matrix often fails to produce accurate approximations when the spectrum decays rapidly.

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D

# === Load data ===
mat_path = "/accounts/postdoc/uid1996247/Research/StructuredRSVD/zeta_testsSS_fullA_updatedRSVD_iid.mat"
data     = scipy.io.loadmat(mat_path)

ks           = data['ks'].flatten()
zeta_vals    = data['zeta_vals'].flatten()
errG_all     = data['errG_all']
errS_all     = data['errS_all']
errI_all     = data['errI_all']
profile_cells = data['profile_names'].squeeze()

if errG_all.ndim == 3: errG_all = errG_all[..., np.newaxis]
if errS_all.ndim == 3: errS_all = errS_all[..., np.newaxis]
if errI_all.ndim == 3: errI_all = errI_all[..., np.newaxis]

if errG_all.shape[1:3] == (len(ks), len(zeta_vals)):
    errG_all = np.transpose(errG_all, (0, 2, 1, 3))
    errS_all = np.transpose(errS_all, (0, 2, 1, 3))
    errI_all = np.transpose(errI_all, (0, 2, 1, 3))

profile_names = []
for raw in profile_cells:
    if isinstance(raw, np.ndarray):
        flat = raw.flatten()
        if flat.dtype.kind in ('U', 'S'):
            name = ''.join(flat.tolist())
        else:
            name = ''.join(chr(int(c)) for c in flat if int(c) != 0)
    else:
        name = str(raw)
    profile_names.append(name)

included_profiles = ["PolyDecaySlow", "PolyDecayMed", "ExpDecaySlow", "ExpDecayMed"]
profile_indices   = [i for i, name in enumerate(profile_names) if name in included_profiles]

numZeta, numKs, numRuns = errG_all.shape[1:]
numProfiles = len(profile_indices)

med_gauss = np.zeros((numProfiles, numKs))
q10_gauss = np.zeros((numProfiles, numKs))
q90_gauss = np.zeros((numProfiles, numKs))

med_subcol = np.zeros((numProfiles, numKs, numZeta))
q10_subcol = np.zeros((numProfiles, numKs, numZeta))
q90_subcol = np.zeros((numProfiles, numKs, numZeta))

med_iid = np.zeros((numProfiles, numKs, numZeta))
q10_iid = np.zeros((numProfiles, numKs, numZeta))
q90_iid = np.zeros((numProfiles, numKs, numZeta))

for pi, p in enumerate(profile_indices):
    for ki in range(numKs):
        g = errG_all[p, 0, ki, :]
        med_gauss[pi, ki] = np.median(g)
        q10_gauss[pi, ki] = np.percentile(g, 10)
        q90_gauss[pi, ki] = np.percentile(g, 90)
        for iz in range(numZeta):
            s = errS_all[p, iz, ki, :]
            med_subcol[pi, ki, iz] = np.median(s)
            q10_subcol[pi, ki, iz] = np.percentile(s, 10)
            q90_subcol[pi, ki, iz] = np.percentile(s, 90)
            i = errI_all[p, iz, ki, :]
            med_iid[pi, ki, iz] = np.median(i)
            q10_iid[pi, ki, iz] = np.percentile(i, 10)
            q90_iid[pi, ki, iz] = np.percentile(i, 90)

# === Plotting ===
eps = 1e-16
gauss_c      = 'black'
cmap_sparse  = matplotlib.colormaps["winter"]
zeta_colors  = cmap_sparse(np.linspace(0.4, 0.8, numZeta))
cmap_iid     = matplotlib.colormaps["gist_heat"]
iid_colors   = cmap_iid(np.linspace(0.4, 0.8, numZeta))

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for pi, ax in enumerate(axs):
    # Gaussian
    lowG, highG = q10_gauss[pi] + eps, q90_gauss[pi] + eps
    medG = med_gauss[pi] + eps
    ax.fill_between(ks, lowG, highG, color=gauss_c, alpha=0.15)
    ax.plot(ks, medG, '-o', color=gauss_c,
            markerfacecolor='none', markeredgewidth=1.5, markersize=12,
            linewidth=1.5)

    # Sparse and IID
    for iz, ζ in enumerate(zeta_vals):
        c_sub = zeta_colors[iz]
        c_iid = iid_colors[iz]
        
        lowI, highI = q10_iid[pi, :, iz] + eps, q90_iid[pi, :, iz] + eps
        medI = med_iid[pi, :, iz] + eps
        ax.fill_between(ks, lowI, highI, color=c_iid, alpha=0.05)
        ax.plot(ks, medI, '-.s', color=c_iid, linewidth=1.5, markersize=6, zorder=4)

        lowS, highS = q10_subcol[pi, :, iz] + eps, q90_subcol[pi, :, iz] + eps
        medS = med_subcol[pi, :, iz] + eps
        ax.fill_between(ks, lowS, highS, color=c_sub, alpha=0.15)
        ax.plot(ks, medS, '--o', color=c_sub, linewidth=1.5, markersize=6, zorder=5)

    ax.set_yscale('log')
    ax.set_title(included_profiles[pi], fontsize=16, family='monospace')
    ax.set_xlabel(r'Sketch dimension $k$', fontsize=14)
    ax.set_ylabel('Relative error', fontsize=14)
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    ax.grid(False)

legend_dict = {
    'Gaussian'       : Line2D([0], [0], color=gauss_c, marker='o', linestyle='-',
                              markerfacecolor='none', markeredgewidth=1.5, markersize=8),
    ' '              : Line2D([], [], linestyle=''),  # spacer
}
for iz, ζ in enumerate(zeta_vals):
    legend_dict[rf'SparseStack $\zeta$={ζ}'] = Line2D([0], [0], color=zeta_colors[iz],
                                               marker='o', linestyle='--',
                                               linewidth=1.5, markersize=6)
for iz, ζ in enumerate(zeta_vals):
    legend_dict[rf'SparseIID $\zeta$={ζ}'] = Line2D([0], [0], color=iid_colors[iz],
                                             marker='s', linestyle='-.',
                                             linewidth=1.5, markersize=6)

handles = list(legend_dict.values())
labels  = list(legend_dict.keys())

plt.tight_layout()
fig.subplots_adjust(top=0.88)  # leave space for legend
fig.legend(handles, labels,
           loc='upper center', bbox_to_anchor=(.533, 1.0),
           ncol=5, fontsize=14, fancybox=True, shadow=False)

plt.savefig("GN_SS_RSVD_selectedProfiles.png", dpi=300, bbox_inches="tight")
plt.show()
```
