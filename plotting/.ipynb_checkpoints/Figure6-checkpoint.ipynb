### Figure X. Fast trigonometric transform test matrices: Comparison.  
Relative error  
$$
\frac{\|\mathbf{A} - \hat{\mathbf{A}}\|_{\mathrm{F}}^2}{\|\mathbf{A}\|_{\mathrm{F}}^2}
$$  
for the rank-$k$ approximation $\hat{\mathbf{A}}$ obtained by the RSVD.  

Sketching is performed with a fast transform test matrix with Rademacher $\mathbf{D}$ and DCT $\mathbf{F}$;  
the sampling matrix $\mathbf{S}$ is either SparseIID (*left*) or SparseCol (*right*).  

All plots compare against a Gaussian test matrix and a fast transform test matrix where $\mathbf{S}$ is subsampling or CountSketch.  

We consider four diagonal input matrices $\mathbf{A} \in \mathbb{R}^{1024 \times 1024}$ from the testbed in Tropp et al. (2017).  
For each instance, we perform 100 trials. The data markers track the median errors, and shaded regions are bounded by the 10% and 90% quantiles.  

On these examples, the SparseCol matrix achieves better accuracy than the SparseIID matrix for every value of the column sparsity $\ell$.

---

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D

# === Load data ===
mat_path      = "/accounts/postdoc/uid1996247/Research/StructuredRSVD/stt_sparsity_LRA_varyZeta_fullA.mat"
data          = scipy.io.loadmat(mat_path, squeeze_me=True, struct_as_record=False)

ks            = data['ks']
zeta_vals     = np.atleast_1d(data['zeta_vals']).flatten()
profile_names = [str(p) for p in data['profile_names'].flatten()]
sketches      = [str(s) for s in data['sketch_types'].flatten()]
err_all       = data['err_all']  # (P, K, S, Z, R)

base_colors       = {
    'gaussian':    'black',
    'uniform':     'black',
    'countsketch': 'violet',
}
markers           = {
    'gaussian':    'o',
    'uniform':     's',       
    'countsketch': 'o',
    'sparsecol':   'o',
    'iid':         's',
}
linestyle_solid   = '-'
linestyle_dashed  = '--'
linestyle_dotdash = '-.'     
tick_size         = 14
font_size         = 14
title_size        = 18
marker_size       = 6
marker_size_gauss = 12
gaussian_lw       = 1.5
gaussian_mew      = 1.5

# === Colormap sampling ===
n_zeta        = len(zeta_vals)
cmap_sparse   = matplotlib.colormaps["winter"]
sparse_colors = cmap_sparse(np.linspace(0.4, 0.8, n_zeta))
cmap_iid      = matplotlib.colormaps["gist_heat"]
iid_colors    = cmap_iid(np.linspace(0.4, 0.8, n_zeta))

# === Compute stats ===
med = np.median(err_all, axis=-1)
q10 = np.percentile(err_all, 10, axis=-1)
q90 = np.percentile(err_all, 90, axis=-1)
eps = 1e-16

# === Method metadata ===
method_meta  = []
method_names = []
for si, name in enumerate(sketches):
    lname = name.lower()
    if lname in ('sparsecol', 'iid'):
        for zi in range(n_zeta):
            method_meta.append((si, zi))
            method_names.append(f"{lname}_{zi+1}")
    else:
        method_meta.append((si, 0))
        method_names.append(lname)

row_profiles = ["PolyDecaySlow", "ExpDecayMed"]
row_idxs     = [profile_names.index(p) for p in row_profiles]

# === M ain plotting loop ===
fig, axes   = plt.subplots(2, 2, figsize=(12, 10), dpi=300)
handle_dict = {}

for row, pidx in enumerate(row_idxs):
    for col, focus in enumerate(('iid', 'sparsecol')):
        ax = axes[row, col]
        for m_idx, (si, zi) in enumerate(method_meta):
            base = method_names[m_idx].split('_')[0]
            show = base in ('gaussian', 'uniform', 'countsketch', focus)
            if not show:
                continue

            med_curve = med[pidx, :, si, zi] + eps

            if base == 'gaussian':
                ln, = ax.plot(
                    ks,
                    med_curve,
                    linestyle_solid,
                    marker=markers['gaussian'],
                    markersize=marker_size_gauss,
                    markerfacecolor='none',
                    markeredgecolor=base_colors['gaussian'],
                    markeredgewidth=gaussian_mew,
                    linewidth=gaussian_lw,
                    color=base_colors['gaussian']
                )

            elif base == 'uniform':
                ax.fill_between(
                    ks,
                    q10[pidx, :, si, zi] + eps,
                    q90[pidx, :, si, zi] + eps,
                    color=base_colors['uniform'],
                    alpha=0.08,
                )
                ln, = ax.plot(
                    ks,
                    med_curve,
                    linestyle_solid,
                    marker=markers['uniform'],
                    markersize=10,
                    markerfacecolor='none',
                    markeredgecolor=base_colors['uniform'],
                    markeredgewidth=gaussian_mew,
                    linewidth=gaussian_lw,
                    color=base_colors['uniform']
                )

            else:
                if base == 'countsketch':
                    color, mface = base_colors['countsketch'], base_colors['countsketch']
                    ls, msize, mew, lw = linestyle_solid, marker_size_gauss - 4, 1, 1.5
                elif base == 'sparsecol':
                    color, mface = sparse_colors[zi], sparse_colors[zi]
                    ls, msize, mew, lw = linestyle_dashed, marker_size, 1, 1.5
                else:  # iid
                    color, mface = iid_colors[zi], iid_colors[zi]
                    ls, msize, mew, lw = linestyle_dotdash, marker_size, 1, 1.5

                ax.fill_between(
                    ks,
                    q10[pidx, :, si, zi] + eps,
                    q90[pidx, :, si, zi] + eps,
                    color=color,
                    alpha=0.15,
                )
                ln, = ax.plot(
                    ks,
                    med_curve,
                    linestyle=ls,
                    marker=markers[base],
                    markersize=msize,
                    markerfacecolor=mface,
                    markeredgecolor=color,
                    markeredgewidth=mew,
                    linewidth=lw,
                    color=color,
                )

            key = f"{base}_{zi+1}" if base in ('sparsecol', 'iid') else base
            handle_dict.setdefault(key, ln)

        ax.set_yscale('log')
        suffix = '-SparseIID' if focus == 'iid' else '-SparseCol'
        title_str = rf"$\mathtt{{{profile_names[pidx]}}}$" + suffix
        ax.set_title(title_str, fontsize=title_size)
        # ax.set_title(profile_names[pidx], fontsize=title_size, family='monospace')
        ax.set_xlabel(r'Sketch dimension $k$', fontsize=font_size)
        ax.set_ylabel('Relative error', fontsize=font_size)
        ax.tick_params(labelsize=tick_size)
        ax.grid(False)

blank          = Line2D([], [], linestyle='')
legend_handles = []
legend_labels  = []
legend_entries = [
    ('gaussian',    'Gaussian'),
    (None,          ''),
    ('uniform',     'Subsampling'),
    ('countsketch', 'CountSketch'),
    ('sparsecol_1', r'SparseCol $\xi=1$'),
    ('sparsecol_2', r'SparseCol $\xi=2$'),
    ('sparsecol_3', r'SparseCol $\xi=3$'),
    ('sparsecol_4', r'SparseCol $\xi=4$'),
    ('iid_1',       r'SparseIID $\xi=1$'),
    ('iid_2',       r'SparseIID $\xi=2$'),
    ('iid_3',       r'SparseIID $\xi=3$'),
    ('iid_4',       r'SparseIID $\xi=4$'),
]
for key, label in legend_entries:
    if key is None:
        legend_handles.append(blank)
    elif key.startswith('iid_'):
        zi = int(key.split('_')[1]) - 1
        legend_handles.append(
            Line2D([], [],
                   linestyle=linestyle_dotdash,
                   marker=markers['iid'],
                   markersize=marker_size,
                   markerfacecolor=iid_colors[zi],
                   markeredgecolor=iid_colors[zi],
                   markeredgewidth=1,
                   color=iid_colors[zi],
                   linewidth=1.5)
        )
    else:
        legend_handles.append(handle_dict[key])
    legend_labels.append(label)

fig.legend(
    legend_handles,
    legend_labels,
    loc='upper center',
    bbox_to_anchor=(0.52, 1.05),
    ncol=6,
    fontsize=14,
    frameon=True,
    framealpha=1,
    shadow=False,
    borderpad=0.7,
    labelspacing=0.3,
    columnspacing=1.0,
    handletextpad=0.5,
    markerscale=1,
)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig(
    "stt_2x2_gauss_vs_sparse_iid.png",
    dpi=300,
    bbox_inches="tight",
)
plt.show()
```
