### Figure X. SparseStack: Scientific data compression. 
Relative error $\|\mathbf{A} - \hat{\mathbf{A}}\|_{\mathrm{F}} / \|\mathbf{A}\|_{\mathrm{F}}$ (*bottom left*),  
runtime (*top left*), and speed-up (*right*) for compression of Gross–Pitaevskii equation (GPE) simulation data of size $131{,}072 \times 40{,}000$ using  
the generalized Nyström method (GN, orange) or the generalized Nyström SVD (GNSVD, blue) with either Gaussian test matrices (solid circles) or SparseStack test matrices (dashed squares).  

We report the median of five runs.

---

```python
import scipy.io
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.colors import to_rgb
import numpy as np

def lighten(color, amount=0.25):
    r, g, b = to_rgb(color)
    return (r + (1 - r) * amount,
            g + (1 - g) * amount,
            b + (1 - b) * amount)

ms_err_gauss = 14
ms_err_ss     = 8
ms_time       = 8

label_fontsize  = 16
tick_fontsize   = 16
legend_fontsize = 16
title_fontsize  = 18

lw    = 2
alpha = 0.9

# --- Load data --------------------------------------------------------------
mat = scipy.io.loadmat(
    'data/POD_multiRun_results.mat',
    struct_as_record=False, squeeze_me=True
)
k = mat['modeList'].flatten()

methods = {
    'GNSVD': ('errs_gnsvd_gauss', 'errs_gnsvd_ss',
              'times_gnsvd_gauss', 'times_gnsvd_ss', 'tab:blue'),
    'GN':    ('errs_gngauss',    'errs_gnss',
              'times_gngauss',   'times_gnss',    'tab:orange'),
}
dash_patterns = {'GNSVD': None, 'GN': (5,1)}

fig = plt.figure(figsize=(12, 8), dpi=300)
gs  = fig.add_gridspec(2, 2, width_ratios=[1, 1], height_ratios=[1, 1], wspace=0.3, hspace=0.3)

ax_time  = fig.add_subplot(gs[0, 0])
ax_err   = fig.add_subplot(gs[1, 0], sharex=ax_time)
ax_speed = fig.add_subplot(gs[:, 1])

# --- Plotting (unchanged) -------------------------------------------------
for name, (e_gauss, e_ss, t_gauss, t_ss, col) in methods.items():
    err_gauss_runs = np.atleast_2d(mat[e_gauss])
    err_ss_runs    = np.atleast_2d(mat[e_ss])
    time_gauss_runs = np.atleast_2d(mat[t_gauss])
    time_ss_runs    = np.atleast_2d(mat[t_ss])

    med_err_gauss = np.median(err_gauss_runs, axis=0)
    q10_err_gauss, q90_err_gauss = np.quantile(err_gauss_runs, [0.1, 0.9], axis=0)
    med_err_ss    = np.median(err_ss_runs, axis=0)
    q10_err_ss, q90_err_ss = np.quantile(err_ss_runs, [0.1, 0.9], axis=0)

    med_time_gauss = np.median(time_gauss_runs, axis=0)
    q10_time_gauss, q90_time_gauss = np.quantile(time_gauss_runs, [0.1, 0.9], axis=0)
    med_time_ss    = np.median(time_ss_runs, axis=0)
    q10_time_ss, q90_time_ss = np.quantile(time_ss_runs, [0.1, 0.9], axis=0)

    # TIME panel 
    ax_time.semilogy(k, med_time_gauss, 'o-', label=f'{name} – Gaussian', color=col,
                     lw=lw, ms=ms_time, mfc='none', alpha=alpha)
    ax_time.fill_between(k, q10_time_gauss, q90_time_gauss, color=lighten(col, 0.6), alpha=0.3)
    ln_ss_t = ax_time.semilogy(k, med_time_ss, 's--', label=f'{name} – SparseStack', color=col,
                               lw=lw, ms=ms_time, alpha=alpha)[0]
    if dash_patterns[name]:
        ln_ss_t.set_dashes(dash_patterns[name])
    ax_time.fill_between(k, q10_time_ss, q90_time_ss, color=lighten(col, 0.6), alpha=0.3)

    # ERROR panel
    ax_err.semilogy(k, med_err_gauss, 'o-', label=f'{name} – Gaussian', color=col,
                    lw=lw, ms=ms_err_gauss+2, mew=1.5, mfc='none', alpha=alpha)
    ax_err.fill_between(k, q10_err_gauss, q90_err_gauss, color=lighten(col, 0.6), alpha=0.3)
    ln_ss = ax_err.semilogy(k, med_err_ss, 's--', label=f'{name} – SparseStack', color=col,
                            lw=lw, ms=ms_err_ss, alpha=alpha)[0]
    if dash_patterns[name]:
        ln_ss.set_dashes(dash_patterns[name])
    ax_err.fill_between(k, q10_err_ss, q90_err_ss, color=lighten(col, 0.6), alpha=0.3)

    # SPEED-UP panel 
    speedup = time_gauss_runs / time_ss_runs
    med_sp   = np.median(speedup, axis=0)
    q10_sp, q90_sp = np.quantile(speedup, [0.1, 0.9], axis=0)
    ax_speed.plot(k, med_sp, 's--', color=col, lw=lw, ms=ms_time,
                  markerfacecolor=col, alpha=alpha, label=name)
    ax_speed.fill_between(k, q10_sp, q90_sp, color=col, alpha=0.3)

for ax, ylabel in zip((ax_time, ax_err, ax_speed),
                      ('Time (s)', 'Relative error', 'Speed-up')):
    ax.set_xscale('log')
    ax.grid(True, which='both', ls='--', lw=0.5)
    ax.set_xlabel(r'Sketch dimension $k$', fontsize=label_fontsize)
    ax.set_ylabel(ylabel, fontsize=label_fontsize)
    ax.tick_params(labelsize=tick_fontsize)

handles = []
for name, (_,_,_,_,col) in methods.items():
    handles.append(Line2D([0],[0], color=col, marker='o', linestyle='-',
                          lw=lw, ms=ms_err_gauss, mfc='none', alpha=alpha,
                          label=f'{name} – Gaussian'))
    handles.append(Line2D([0],[0], color=col, marker='s', linestyle='--',
                          lw=lw, ms=ms_err_ss, alpha=alpha,
                          label=f'{name} – SparseStack'))

fig.legend(handles=handles, ncol=2, loc='upper center', bbox_to_anchor=(0.5, 1.02),
           fancybox=True, shadow=False, fontsize=legend_fontsize, frameon=True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
fig.savefig('fig9.jpg', bbox_inches='tight')
plt.show()
```
